```python
import numpy as np
import matplotlib.pyplot as plt
import random
print("all good")
```

```python
portfolio_value = []

portfolio = 100
days = 50

for i in range(days):
    profit_loss = random.choice([-1,1])
    portfolio = portfolio + profit_loss
    portfolio_value.append(portfolio)
print(portfolio_value)

print(" ")

print("Highest Portfolio Value:",max(portfolio_value))
print("Lowest Portfolio Value:",min(portfolio_value))
print("Final Portfolio Value:",portfolio_value[-1])

print(" ")

plt.plot(portfolio_value)
plt.xlabel("Time")
plt.ylabel("Portfolio Value")
plt.title("Value of Portfolio Over 50 Days")
plt.show()



    
```
