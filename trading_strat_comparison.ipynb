```python
import matplotlib.pyplot as plt
import random
print("all good")
```

```python
portfolio_A = []
portfolio_B = []

trader_A = 100
trader_B = 100
days = 100

for i in range(days):
    gain_loss_A = random.choice([-1,1])
    gain_loss_B = random.choice([-1,-1,1,1,1])
    trader_A = trader_A + gain_loss_A
    trader_B = trader_B + gain_loss_B
    portfolio_A.append(trader_A)
    portfolio_B.append(trader_B)
    
print(portfolio_A)

print("")

print("Highest Value:",max(portfolio_A))
print("Lowest Value:",min(portfolio_A))
print("Final Value:",portfolio_A[-1])

print("")

print(portfolio_B)

print("")

print("Highest Value:",max(portfolio_B))
print("Lowest Value:",min(portfolio_B))
print("Final Value:",portfolio_B[-1])

print("")

plt.plot(portfolio_A)
plt.plot(portfolio_B)
plt.xlabel("Time")
plt.ylabel("Portfolio Value")
plt.title("Trader A (Blue) against Trader B (Orange)")
plt.show()


```
