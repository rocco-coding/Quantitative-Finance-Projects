```python
positions = []

num_steps = 100
current_position = 0

for i in range(num_steps):
   step = random.choice([-1,1])
   current_position = current_position + step
   positions.append(current_position)
    #print(current_position)

print(positions)

```

```python
highest_position = max(positions)
lowest_position = min(positions)

print("Highest:",highest_position)
print("Lowest:",lowest_position)
print("Final:",positions[-1])
    

```

```python
plt.plot(positions)
plt.xlabel("Step Number")
plt.ylabel("Position")
plt.title("First Random Walk")
plt.show()
```
