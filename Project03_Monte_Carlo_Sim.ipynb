```python
# Project 03: Monte Carlo Simulator

# Question:
# What range of final outcomes are possible under simulated market conditions?
```

```python
import matplotlib.pyplot as plt
import random
print('all good')
```

```python
final_values = []
portfolio_num = 100
days = 100

for i in range(portfolio_num):
    equity_per = 100
    for i in range(days):
        profit_loss = random.choice([-1,1])
        equity_per = equity_per + profit_loss
    final_values.append(equity_per)
print(final_values)
print('')
print('Highest Value:',max(final_values))
print('Lowest Value:',min(final_values))
print('Average Value:',sum(final_values)/portfolio_num)
print('')

plt.hist(final_values)
plt.xlabel('Portfolio Value')
plt.ylabel('Number of Portfolios')
plt.title('Values of 100 Portfolios after 100 days')
plt.show()
```

```python
# Conclusion

# The Monte Carlo Simulation generated many 
# possible portflio paths and produced a distribution
# of final portfolio values.

# The histogram showed the range, frequency and likelihood of different outcomes,
# providing insight into both expected returns and investment uncertainty.
```
