```python
# Project 04: Monte Carlo Drawdown Analysis

# Question:
# What is the distribution of maximum drawdowns across
# simulated portfolio paths?
```

```python
import matplotlib.pyplot as plt
import random
```

```python
portfolio_num = 100
trading_days = 250
max_drawdowns = []

for i in range(portfolio_num):
    portfolio = 100
    portfolio_values = []
    for j in range(trading_days):
        profit_loss = random.choice([-1,1])
        portfolio = portfolio + profit_loss
        portfolio_values.append(portfolio)
    peak = portfolio_values[0]
    max_drawdown = 0

    for value in portfolio_values:
        if value > peak:
            peak = value
        current_drawdown = ((peak - value)/peak)*100
        if current_drawdown > max_drawdown:
            max_drawdown = current_drawdown
    max_drawdowns.append(max_drawdown)

print("Average Max Drawdown:",sum(max_drawdowns)/len(max_drawdowns))
print("Worst Max Drawdown:",max(max_drawdowns))
print("Best Max Drawdown:",min(max_drawdowns))

plt.hist(max_drawdowns)
plt.xlabel("Percentage Drawdown")
plt.ylabel("Number of Portfolios")
plt.title("Max Drawdowns for 100 Portfolios over 250 days")
plt.show()


    
```

```python
# Conclusion

# Simulated portfolios experienced a range of
# maximum drawdowns.

# The histogram shows the distribution
# of drawdown risk and highlights how severe
# losses could be, under different market scenarios.
```
